```python
import sklearn.datasets
import pandas as pd
cancer = sklearn.datasets.load_breast_cancer()

#create a dataframe variable 'a' with this dataset. This dataframe should have all the 569 instances, 30 features and the class of 569 instances as 0 or 1.
a = pd.DataFrame(cancer.data, columns=cancer.feature_names)
#change the class of 569 instances as 0 or 1 to 'malignant' or 'benign'
a['typeofcancer'] = cancer.target
a.shape
```

```python
#slice the dataframe 'a' with three features 'mean radius', 'mean area' and 'mean perimeter'
df = a[['mean radius','mean perimeter','mean area','typeofcancer']]
df.iloc[0:2]
```

```python
#show rows with indicies 17, 18, 19, 20, 21
df.iloc[17:22]
```

TASK 2

```python
#generate hollowing figure with 3 subplots. 
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,3, figsize=(15,5))
#create labels for the three subplots
ax[0].set_xlabel('mean radius')
ax[0].set_ylabel('frequency')
ax[1].set_xlabel('mean perimeter')
ax[1].set_ylabel('mean radius')
ax[2].set_xlabel('mean area')
ax[2].set_ylabel('mean radius')
ax[2].semilogy()

#histogram of ‘mean radius’ data for each class,
ax[0].hist(df[df['typeofcancer']==0]['mean radius'], alpha=0.5, label='benign', color='red')
ax[0].hist(df[df['typeofcancer']==1]['mean radius'], alpha=0.5, label='malignant', color='blue')
#scatter plot of ‘mean radius’ vs ‘mean perimeter for each class, range of y-axis should be 0 to 30 increasing in increments of 5, range of x-axis should be 0 to 200, increasing in increments of 50
ax[1].scatter(df[df['typeofcancer']==0]['mean radius'], df[df['typeofcancer']==0]['mean perimeter'], label='benign', color='blue')
ax[1].scatter(df[df['typeofcancer']==1]['mean radius'], df[df['typeofcancer']==1]['mean perimeter'], label='malignant', color='orange')
#scatter plot of ‘mean radius’ vs ‘mean area’ for each class

ax[2].scatter(df[df['typeofcancer']==0]['mean radius'], df[df['typeofcancer']==0]['mean area'], label='benign', color='blue')
ax[2].scatter(df[df['typeofcancer']==1]['mean radius'], df[df['typeofcancer']==1]['mean area'], label='malignant', color='orange')
```

The first subplot reveals that cancer patients with benign cancer status tend to have a larger mean radius than patients with malignant cancer status. The second subplot reveals that cancer patients with benign cancer status tend to have a smaller mean perimeter than patients with malignant cancer status. The third subplot reveals that cancer patients with benign cancer status tend to have a smaller mean area than patients with malignant cancer status.

```python


#import Car_sales.csv file as pandas dataframe
data = pd.read_csv('Car_sales.csv')
#print the first 5 rows of the dataframe
#data.head()
df = data[['Manufacturer','Horsepower','Price_in_thousands','Sales_in_thousands']]
#create a new dataframe with only the rows where the manufacturer is Ford and Dodge
df2 = df[(df['Manufacturer']=='Ford') | (df['Manufacturer']=='Dodge')]
#print the first 5 rows of the dataframe
df2.head()


```
